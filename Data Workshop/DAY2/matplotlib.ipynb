```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
data = {
    "A" : [1,2,3,4,5,6,7],
    "B" : [10,20,30,40,50,60,70]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize = (10,6))
plt.bar(df.index, df["B"], color = 'orange')
plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data1 = {
    "nilu" : [1000,2000,3000,4000,5000,6000,7000],
    "prapti" : [1550,5520,5430,4560,5870,6340,9870]
}
```

```python
df1 = pd.DataFrame(data1)
df1.head()
```

```python
#line plot
plt.figure(figsize = (10,6))
plt.plot(df1.index, df1["nilu"], label = 'nilu',marker ='o')
plt.plot(df1.index, df1["prapti"],label = 'prapti',marker ='x')
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

plt.show()
```
